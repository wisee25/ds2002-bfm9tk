## Mapping with Folium

```python
map_airbnb = folium.Map(location=[40.7, -73.85], zoom_starts=12)

for _, row in airbnb.iterrows():
    folium.Marker(
        location = [row['Latitude'], row['Longitude']],
        popup=f"Price: ${row['Price']} | 'Rating': {row['Rating']}:.1f"
    ).add_to(map_airbnb)
    
map_airbnb
```
